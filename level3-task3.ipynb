```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")
df.dropna(subset=['Cuisines'], inplace=True)

plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='purple')
plt.title("Distribution of Aggregate Ratings")
plt.tight_layout()
plt.show()

df_cuisine_avg = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=df_cuisine_avg, x='Aggregate rating', y='Cuisines', palette='mako')
plt.title("Top 10 Cuisines by Average Rating")
plt.tight_layout()
plt.show()

df_city_avg = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=df_city_avg, x='Aggregate rating', y='City', palette='coolwarm')
plt.title("Top 10 Cities by Average Rating")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Votes', y='Aggregate rating', hue='Price range', palette='husl')
plt.title("Votes vs Aggregate Rating (colored by Price Range)")
plt.tight_layout()
plt.show()
```
