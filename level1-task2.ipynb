```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Dataset .csv")

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='viridis')
plt.title("Feature Correlation Matrix")
plt.show()

```

```python

plt.figure(figsize=(8,5))
sns.boxplot(x='Price range', y='Aggregate rating', data=df)
plt.title("Price Range vs Aggregate Rating")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Aggregate rating', y='Votes', hue='Price range', palette='cool')
plt.title("Votes vs Rating (Colored by Price Range)")
plt.show()
```

```python
top_cities = df.groupby('City')['Votes'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_cities.plot(kind='bar', color='purple')
plt.title("Top Cities by Total Votes")
plt.ylabel("Total Votes")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.show()
```
