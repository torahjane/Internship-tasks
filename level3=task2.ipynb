```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")
df.dropna(subset=['Cuisines'], inplace=True)

df_grouped = df.groupby('Cuisines').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Restaurant Name': 'count'
}).reset_index()

df_grouped.columns = ['Cuisines', 'Average Rating', 'Total Votes', 'Number of Restaurants']
df_top_votes = df_grouped.sort_values(by='Total Votes', ascending=False).head(10)
df_top_rating = df_grouped[df_grouped['Number of Restaurants'] > 5].sort_values(by='Average Rating', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=df_top_votes, x='Total Votes', y='Cuisines', palette='plasma')
plt.title("Top 10 Most Voted Cuisines")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(data=df_top_rating, x='Average Rating', y='Cuisines', palette='viridis')
plt.title("Top 10 Highest Rated Cuisines (with >5 Restaurants)")
plt.tight_layout()
plt.show()
```
