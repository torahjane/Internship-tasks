```python
import pandas as pd

df = pd.read_csv("Dataset .csv")
df.head()
```

```python
df['Name Length'] = df['Restaurant Name'].apply(lambda x: len(str(x)))
df['Address Length'] = df['Address'].apply(lambda x: len(str(x)))

df['Table_Booking_Flag'] = df['Has Table booking'].apply(lambda x: 1 if x.strip().lower() == 'yes' else 0)
df['Online_Delivery_Flag'] = df['Has Online delivery'].apply(lambda x: 1 if x.strip().lower() == 'yes' else 0)

df[['Restaurant Name', 'Name Length', 'Address Length', 'Table_Booking_Flag', 'Online_Delivery_Flag']].head()
```

```python
# Average rating by table booking flag
df.groupby('Table_Booking_Flag')['Aggregate rating'].mean()
# Average rating by name length group
df['Name Length Group'] = pd.cut(df['Name Length'], bins=[0, 10, 20, 30, 50], labels=['Short', 'Medium', 'Long', 'Very Long'])
df.groupby('Name Length Group')['Aggregate rating'].mean()
```
