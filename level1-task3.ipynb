```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Dataset .csv")

df = df.dropna(subset=['Latitude', 'Longitude'])
country_counts = df['Country Code'].value_counts().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='viridis')
plt.title("Top Countries by Restaurant Count")
plt.xlabel("Number of Restaurants")
plt.ylabel("Country Code")
plt.show()

```

```python
top_rated = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
top_rated.plot(kind='barh', color='teal')
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("City")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Aggregate rating'], cmap='viridis', alpha=0.6)
plt.colorbar(label='Aggregate Rating')
plt.title("Restaurant Ratings by Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
